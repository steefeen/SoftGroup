```python
import numpy as np
from tqdm.notebook import tqdm
import torch
import os
from dgl.geometry import farthest_point_sampler

input_path = '/content/drive/MyDrive/Hauptseminar/data/stpls3d/val'
output_path = '/content/data/stpls3d/manipulated_data'
path_list = [f'{input_path}/{f}' for f in os.listdir(input_path) if f.endswith('.pth')]


def remove_color(data, _=None):
    data[1] = np.zeros(data[1].shape)
    return data

def add_noise_to_cords(data, s):
    data[0] = np.random.normal(data[0], s)
    return data

def sample_farthest(data, p):
    n_new = int(len(data[0])*p)
    arr = np.array([data[0]])
    point_idx = farthest_point_sampler(torch.from_numpy(arr), n_new)
    return [a[point_idx] for a in data]

def sample_random(data, p):
    mask = np.random.rand(len(data[0])) < p
    return [a[mask] for a in data]

def manipulate_data(f, name: str, values:list=None):
    values = values if values else ['']
    for value in tqdm(values):
        out_path = f'{output_path}/{name}-{value}/val'
        os.makedirs(out_path, exist_ok=True)
        f_adapted = lambda x: f(x, value)
        
        for file_path in tqdm(path_list, leave=False):
            file_name = file_path.split('/')[-1]
            data = torch.load(file_path)
            data = tuple(f_adapted(list(data)))
            torch.save(data, f'{out_path}/{file_name}')
```

```python
manipulate_data(sample_random, 'sample_random', [1.0, 0.95, 0.9, 0.85, 0.8, 0.7])
manipulate_data(remove_color, 'no_color', [0])
manipulate_data(add_noise_to_cords, 'noisy', [0, 0.1, 0.2, 0.5, 0.8, 1.0])
#manipulate_data(sample_farthest, 'sample_farthest', [0.95, 0.9, 0.85, 0.8, 0.7])
#manipulate_data(sample, 'sample', [1.0, 0.95])
```

```python
class Arguments():
    def __init__(self):
        self.config = 'configs/softgroup_stpls3d.yaml'
        self.checkpoint = 'softgroup_stpls3d_spconv2.pth'
        self.out = None #"./results"
        self.dist = False
```

```python
from glob import glob
import os
import shutil
import time
import logging
from softgroup.util import get_root_logger
import subprocess
import argparse
#from tools.test import evaluate
from tools.test import main as evaluate

logger = get_root_logger()
logger.setLevel("DEBUG")

data_paths = glob("/content/data/stpls3d/manipulated_data/*/", recursive = False)

for source_dir in data_paths:

    
    #logger.info(source_dir)
    
    name = source_dir.split(os.sep)[-2]

    logger.handlers.clear()
    splits = name.split("-")
    os.makedirs(f"/content/drive/MyDrive/Hauptseminar/log/{splits[-2]}/", exist_ok=True)
    print("split", splits[-1])
    handler = logging.FileHandler(f"/content/drive/MyDrive/Hauptseminar/log/{splits[-2]}/{splits[-1]}.log")
    logger.addHandler(handler)
    
    logger.info(name)

    target_dir = '/content/dataset/stpls3d/val'

    print("delete and create folder: ", target_dir)
    if os.path.exists(target_dir):
        shutil.rmtree(target_dir)
    os.mkdir(target_dir)

    time.sleep(2)
    print("move files from: ", source_dir, )
    source_dir = source_dir + "val"
    file_names = os.listdir(source_dir)

    for file_name in file_names:
        #print(source_dir + os.sep + file_name)
        shutil.move(source_dir + os.sep +file_name, target_dir)#copy
    
    print("moved files: ", len(file_names))

    time.sleep(2)
    print("evaluate model")
    results_folder = "./results_manipulated/" + name + "/results_stpls3d"
    print(results_folder)  

    evaluate(args = Arguments())

    time.sleep(2)
    print("delete folder: ", target_dir)
    #shutil.rmtree(target_dir)
print("end")
```
